```python
def error(x, y, r):
    return x**2 + y**2 - r**2

def half(x, y, r):
    return error(x-0.5, y, r)

def inside(pre, x, y):
    return pre - 2 * x + 2 * y

def outside(pre, y):
    return pre - 1 + 2 * y

def main(r):
    x = r
    x_circle = [x]
    y_circle = [0]
    for y in range(1, r):
        if y == 1:
            _h = half(x, y, r)
        elif _h > 0:
            x -= 1
            _h = inside(_h, x, y)
        else:
            _h = outside(_h, y)
        x_circle.append(x)
        y_circle.append(y)
        if y >= x:
            break
    return x_circle, y_circle
```

```python
r = 5
x_circle, y_circle = main(r)
```

```python
_temp_y = [i for i in reversed(x_circle)][1:]
_temp_x = [i for i in reversed(y_circle)][1:]
y_circle += _temp_y
x_circle += _temp_x

_temp_x = [-i for i in reversed(x_circle)][1:]
_temp_y = [i for i in reversed(y_circle)][1:]
x_circle += _temp_x
y_circle += _temp_y

_temp_x = [i for i in reversed(x_circle)][1:-1]
_temp_y = [-i for i in reversed(y_circle)][1:-1]
x_circle += _temp_x
y_circle += _temp_y
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(x_circle, y_circle, marker='s')
ax.add_artist(plt.Circle((0,0), r, color='r', fill=False))

ax.set_aspect(1)
ax.set_xlim(-r-1, r+1)
ax.set_ylim(-r-1, r+1)
```
